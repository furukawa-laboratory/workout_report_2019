# 好きな関数を表示(3次元)

```python
%matplotlib nbagg
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3, 3, 0.01)
y = np.arange(-3, 3, 0.01)
X, Y = np.meshgrid(x, y)
print(X,Y)
Z =np.exp(-(X**2+Y**2))

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X,Y,Z)

plt.show()
```

#### シンプルで分かりやすい形をしてるので好き 3次元上ではメッシュの使い方は直感的に理解できた

# アニメーションで関数を動かす

```python
import matplotlib.animation as anm
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML

fig = plt.figure(figsize = (10, 6))
x = np.arange(-10, 10, 0.1)

def update(i, fig_title, A):
    plt.cla()                      # 現在描写されているグラフを消去

    y = A*x**(i*2)
    plt.plot(x, y, "r")
    plt.title(fig_title + 'i=' + str(i))


ani = anm.FuncAnimation(fig, update, fargs = ('Initial Animation! ', 2.0), \
    interval = 1000, frames = 10)
display(HTML(ani.to_html5_video()))
```

#### グラフを動かしているというより、関数を変更させているだけなので題意に沿っていないかも
