自分の好きな関数を3つ選んで同じfigureでplotを用いて描画する.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x1, y1 = [], []

# 円軌道を生成
# linspaceは指定区間に指定個数の等差数列を生み出す
for j in np.linspace(0, 2*np.pi, 1000):
    x1.append(math.sin(j))
    y1.append(math.cos(j))
```

```python
x2, y2 = [], []

for j in np.linspace(-1, 1, 20):
    x2.append(j)
    y2.append(j**3)
```

```python
x3, y3 = [], []

for j in np.linspace(-1, 1, 20):
    x3.append(j)
    y3.append(-j)
```

```python
plt.figure(1, figsize=(5, 5))
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.show()
```

```python
plt.figure(2, figsize=(10, 3))

plt.subplot(131)
plt.plot(x1, y1)
plt.subplot(132)
plt.plot(x2, y2)
plt.subplot(133)
plt.plot(x3, y3)
plt.show()
```

自分の好きな関数を１つ選んで3次元のプロットを実行する
