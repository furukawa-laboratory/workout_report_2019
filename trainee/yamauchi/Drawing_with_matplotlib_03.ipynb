### 自分の好きな関数を3つ選んで同じfigureでplotを用いて描画する.

$$ x^2 * y^2 = 1 $$
$$ y = x^3 $$
$$ y = -x $$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x1, y1 = [], []

# 円軌道を生成
# linspaceは指定区間に指定個数の等差数列を生み出す
for j in np.linspace(0, 2*np.pi, 1000):
    x1.append(math.sin(j))
    y1.append(math.cos(j))
```

```python
x2, y2 = [], []

for j in np.linspace(-1, 1, 20):
    x2.append(j)
    y2.append(j**3)
```

```python
x3, y3 = [], []

for j in np.linspace(-1, 1, 20):
    x3.append(j)
    y3.append(-j)
```

```python
plt.figure(1, figsize=(5, 5))
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.show()
```

```python
plt.figure(2, figsize=(10, 3))

plt.subplot(131)
plt.plot(x1, y1)
plt.subplot(132)
plt.plot(x2, y2)
plt.subplot(133)
plt.plot(x3, y3)
plt.show()
```

### 自分の好きな関数を１つ選んで3次元のプロットを実行する

```python
%matplotlib notebook

from mpl_toolkits.mplot3d import Axes3D

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

ax.set_title("3D Plot")

ax.set_xlabel("x", size = 15)
ax.set_ylabel("y", size = 15)
ax.set_zlabel("z", size = 15)

x = np.linspace(-8, 8, 256)
y = np.linspace(-8, 8, 256)

X, Y = np.meshgrid(x, y)

Z = X**2 - Y**2

ax.plot_wireframe(X, Y, Z, color="darkblue")

plt.show()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

ax.set_title("3D Plot")

ax.set_xlabel("x", size = 15)
ax.set_ylabel("y", size = 15)
ax.set_zlabel("z", size = 15)

x = np.linspace(-8, 8, 256)
y = np.linspace(-8, 8, 256)

u = np.linspace(0, 2*np.pi, 50)
v = np.linspace(-4*np.pi, 4*np.pi, 50)

u, v = np.meshgrid(u, v)
a = 2
b = 8

X = (b + a*np.cos(u)) * np.cos(v)
Y = (b + a*np.cos(u)) * np.sin(v)
Z = a * np.sin(u) + v

ax.plot_wireframe(X, Y, Z, color="darkblue")

plt.show()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

ax.set_title("3D Plot")

ax.set_xlabel("x", size = 15)
ax.set_ylabel("y", size = 15)
ax.set_zlabel("z", size = 15)

x = np.linspace(-8, 8, 256)
y = np.linspace(-8, 8, 256)

u = np.linspace(0, 2*np.pi, 50)
v = np.linspace(-4*np.pi, 4*np.pi, 50)

u, v = np.meshgrid(u, v)
a = 2
b = 5

X = (b + a*np.cos(u)) * np.cos(v)
Y = (b + a*np.cos(u)) * np.sin(v)
Z = a * np.sin(u)

ax.plot_wireframe(X, Y, Z, color="darkblue")

plt.show()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

ax.set_title("3D Plot")

ax.set_xlabel("x", size = 15)
ax.set_ylabel("y", size = 15)
ax.set_zlabel("z", size = 15)

u = np.linspace(0, 2 * np.pi, 50)
v = np.linspace(0, np.pi, 50)

x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

ax.plot_wireframe(x, y, z, color="darkblue")

plt.show()
```

### animationを使って自分の好きな関数を動かす

```python
%matplotlib notebook
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
fig = plt.figure(1, figsize=(5, 3))
ims = []

for i in range(100):
    r = np.linspace(10, 20, 250)
    im = plt.plot(np.sin(np.pi*r + i/5))
    ims.append(im)

ani = animation.ArtistAnimation(fig, ims, interval=100)

plt.show()
```
