## MinMaxScaler : data mapping to specified range

```python
from IPython.core.display import display, HTML
display(HTML('<style>.container{width:90%!important;}</style>'))

import pandas, scipy, numpy
from sklearn.preprocessing import MinMaxScaler

df=pandas.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv ',sep=';')
array=df.values #array.shape(1599, 12)
x=array[:,0:8]  #x.shape(1599, 8)
y=array[:,8]    #y.shape(1599, 1)
scaler=MinMaxScaler(feature_range=(0,1)) #Setting MinMaxScaler as scaler
rescaledX=scaler.fit_transform(x)
numpy.set_printoptions(precision=3) #Setting precision for print
rescaledX[0:5,:]
```

## StandardScaler : data standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(x) #Setting StandardScaler as scaler with fitting shape of x
StandardX=scaler.transform(x)
StandardX[0:5,:]
```

## Normalizer : data normalization

```python
from sklearn.preprocessing import Normalizer
scaler=Normalizer().fit(x)
normalizedX=scaler.transform(x)
```

## Binarizer : data binalization

```python
from sklearn.preprocessing import Binarizer
binarizer=Binarizer(threshold=0.0).fit(x)
binaryX=binarizer.transform(x)
binaryX[0:5,:]
```

## Scale : data scaling (mean = 0, variance = 1)

```python
from sklearn.preprocessing import scale
data_standardized=scale(df)
print(data_standardized.mean(axis=0))
print(data_standardized.var(axis=0))
```

## OneHotEncoder : encode labels to one-hot represent

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder.fit([[0,1,6,2],
             [1,5,3,5],
             [2,4,2,7],
             [1,0,4,2]])

encoder.transform([[2,4,3,4]]).toarray()
```

## LabelEncoder : encoding label to numeric

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
input_classes=['Havells','Philips','Syska','Eveready','Lloyd']
label_encoder.fit(input_classes)

for i,item in enumerate(label_encoder.classes_):
    print(item,'-->',i)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA

iris = datasets.load_iris() #
X = iris.data[:,:2] #only take the first two features
Y = iris.target #target
print(X.shape, Y.shape)

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5 #
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5 #0

plt.figure(5, figsize=(8,6)) #
plt.clf() #clear the current figure 

plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Set2, edgecolor='k') #iris1,2x,y
plt.colorbar() #
plt.xlabel('1st component of iris data')
plt.ylabel('2nd component of iris data')

plt.xlim(x_min, x_max) #
plt.ylim(y_min, y_max)
plt.xticks(()) #ticks
plt.yticks(())

fig = plt.figure(1, figsize=(8,6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:,0],X_reduced[:,1],X_reduced[:,2], c=Y, cmap=plt.cm.Set2, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
