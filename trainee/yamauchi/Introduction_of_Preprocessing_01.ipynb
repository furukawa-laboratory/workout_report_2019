# Data Preprocessing, Analysis & Visualization

## MinMaxScaler : data mapping to specified range

```python
import pandas, scipy, numpy
from sklearn.preprocessing import MinMaxScaler

df=pandas.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv ',sep=';')
array=df.values #array.shape(1599, 12)
x=array[:,0:8]  #x.shape(1599, 8)
y=array[:,8]    #y.shape(1599, 1)
scaler=MinMaxScaler(feature_range=(0,1)) #Setting MinMaxScaler as scaler
rescaledX=scaler.fit_transform(x)
numpy.set_printoptions(precision=3) #Setting precision for print
rescaledX[0:5,:]
```

## StandardScaler : data standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(x) #Setting StandardScaler as scaler with fitting shape of x
StandardX=scaler.transform(x)
StandardX[0:5,:]
```

## Normalizer : data normalization

```python
from sklearn.preprocessing import Normalizer
scaler=Normalizer().fit(x)
normalizedX=scaler.transform(x)
```

## Binarizer : data binalization

```python
from sklearn.preprocessing import Binarizer
binarizer=Binarizer(threshold=0.0).fit(x)
binaryX=binarizer.transform(x)
binaryX[0:5,:]
```

## Scale : data scaling (mean = 0, variance = 1)

```python
from sklearn.preprocessing import scale
data_standardized=scale(df)
print(data_standardized.mean(axis=0))
print(data_standardized.var(axis=0))
```

## OneHotEncoder : encode labels to one-hot represent

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder.fit([[0,1,6,2],
             [1,5,3,5],
             [2,4,2,7],
             [1,0,4,2]])

encoder.transform([[2,4,3,4]]).toarray()
```

## LabelEncoder : encoding label to numeric

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
input_classes=['Havells','Philips','Syska','Eveready','Lloyd']
label_encoder.fit(input_classes)

for i,item in enumerate(label_encoder.classes_):
    print(item,'-->',i)
```

## PCA : 
PCA[[wiki](https://ja.wikipedia.org/wiki/%E4%B8%BB%E6%88%90%E5%88%86%E5%88%86%E6%9E%90)]



```python
from sklearn.decomposition import PCA
DataX1 = numpy.array([[2,3],
                      [4,6],
                      [5,1],
                      [3,8],
                      [6,9]])
pca1 = PCA(n_components=2)
pca1.fit(DataX1)
point1 = pca.transform(DataX1)
pca1.explained_variance_ratio_
```

```python
pca1.get_covariance()
```

```python
pca1.mean_
```

```python
pca1.components_
```

```python
plt.scatter(*point1.T, marker=".", color='red')
plt.xlabel("PC1") 
plt.ylabel("PC2")
plt.show()
```
