## Working with multiple figures and axes

```python
import numpy as np

# 2つのsuplotsを作るスクリプト
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1) # plt.figure(1) = subplot(111)のこと
plt.subplot(2,1,1)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(2,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## Logarithmic and other nonlinear axes

```python
# 非線形な軸のスケールも使うことができる
# 以下に、同じデータを4つの異なる軸のスケールでプロットする
from matplotlib.ticker import NullFormatter # useful for 'logit' scale

# Fixing random state for reproducibility(再現性のため、ランダム状態を固定する)
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
# loc: 平均（最も度数の多い場所）、scale: 分散、size: 出力サイズ
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
# numpy配列の時に使える閾値処理
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# Log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log（対称性のあるlog）
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit（matplotlibに詳細がないため良く分からないscale）
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# このままだとylabelが多すぎる
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# 1-10^{-3}があるためsuplotのレイアウトを調節
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95,
                    hspace=0.35, wspace=0.35)
plt.show()
```

## pcolormesh

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
import numpy as np

# make these smaller to increase the resolution
dx, dy = 0.05, 0.05

# mgrid = meshgrid
y, x = np.mgrid[slice(1, 5 + dy, dy),
                slice(1, 5 + dx, dy)]

z = np.sin(x)**10 + np.cos(10 + y*x) * np.cos(x)

# x and y are bounds, so z should be the value *inside* those bound.
# Therefore, remove the last value from the z array.
z = z[:-1, :-1]
levels = MaxNLocator(nbins=15).tick_values(z.min(), z.max())
# nbins: Maximum number of intervals; one less than max number of ticks. If the string 'auto', the number of bins will be automatically determined based on the length of the axis.
# tick_values(vmin, vmax): Return the values of the located ticks given vmin and vmax
print('levels: {}'.format(levels))

# pick the desired colormap, sensible levels, and define a normalization
# instance which takes data values and translates those into levels.
cmap = plt.get_cmap('PiYG')
norm = BoundaryNorm(levels, ncolors=cmap.N, clip=True)
print('cmap.N: {}'.format(cmap.N))

fig, (ax0, ax1) = plt.subplots(nrows=2)

im = ax0.pcolormesh(x, y, z, cmap=cmap, norm=norm)
fig.colorbar(im, ax=ax0)
ax0.set_title('pcolormesh with levels')

# contours are *point* based plots. so convert our bound into point
# centers
# contourf() は塗りつぶした等高線を作成するので、
# 等高線のみ作成する場合は contour() 関数を使用する。
cf = ax1.contourf(x[:-1, :-1] + dx/2.,
                  y[:-1, :-1] + dy/2., z, levels=levels,
                  cmap = cmap)
fig.colorbar(cf, ax=ax1)
ax1.set_title('contourf with levels')

# adjust spacing between subplots so 'ax1' title and 'ax0' tick labels
# don't overlap
fig.tight_layout()

plt.show()
```

## PathPatch object
https://matplotlib.org/api/path_api.html

```python
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

Path = mpath.Path
path_data = [
    # MOVETO(1 vertex): Pick up the pen and move to the given vertex.
    # CURVE4(2 control points, 1 endpoint): Draw a cubic Bezier curve 
    # from the current position, with the given control points, to the given end point
    # CLOSEPOLY (1 vertex)(ignored):  Draw a line segment to the start point of the current polyline.
    (Path.MOVETO, (1.58, -2.57)),
    (Path.CURVE4, (0.35, -1.1)),
    (Path.CURVE4, (-1.75, 2.0)),
    (Path.CURVE4, (0.375, 2.0)),
    (Path.LINETO, (0.85, 1.15)),
    (Path.CURVE4, (2.2, 3.2)),
    (Path.CURVE4, (3, 0.05)),
    (Path.CURVE4, (2.0, -0.5)),
    (Path.CLOSEPOLY, (1.58, -2.57)),
]

codes, verts = zip(*path_data)
print('codes: {}'.format(codes))
print('verts: {}'.format(verts))
path = mpath.Path(verts, codes)
print(path)
patch = mpatches.PathPatch(path, facecolor='r', alpha=0.5)
ax.add_patch(patch)

# plot control points and connecting lines
x, y = zip(*path.vertices)
print('x: {}'.format(x))
print('y: {}'.format(y))
line, = ax.plot(x, y, 'go-')

ax.grid()
ax.axis('equal')
plt.show()
```

## 3D surface(color map)

```python
# This import registers the 3D projection, but is otherwise unused.
# ちなみにAxes3Dは明示的に使用していないが，importしておかないとKeyError: '3d'がでる． 
from mpl_toolkits.mplot3d import Axes3D #noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
print('type(ax): {}'.format(type(ax)))

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Streamplot

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

w = 3
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U*U + V*V)

fig = plt.figure(figsize=(7, 9))
gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1,1,2])

# Varying density along a streamline
ax0 = fig.add_subplot(gs[0, 0])
ax0.streamplot(X, Y, U, V, density=[0.5, 1])
ax0.set_title('Varying Density')

# Varying color along a streamline
ax1 = fig.add_subplot(gs[0, 1])
strm = ax1.streamplot(X, Y, U, V, color=U, linewidth=2, cmap='autumn')
fig.colorbar(strm.lines)
ax1.set_title('Varying Color')

# Varying line width along a streamline
ax2 = fig.add_subplot(gs[1,0])
lw = 5*speed / speed.max()
ax2.streamplot(X, Y, U, V, density=0.6, color='k', linewidth=lw)
ax2.set_title('Varying Line Width')

# Controlling the starting points of the streamlines
seed_points = np.array([[-2, -1, 0, 1, 2, -1], [-2, -1, 0, 1, 2, 2]])
ax3 = fig.add_subplot(gs[1,1])
strm = ax3.streamplot(X, Y, U, V, color=U, linewidth=2,
                     cmap='autumn', start_points=seed_points.T)
fig.colorbar(strm.lines)
ax3.set_title('Controlling Starting Points')

# Displaying the starting points with blue symbols
ax3.plot(seed_points[0], seed_points[1], 'bo')
ax3.axis((-w, w, -w, w))

# Create a mask
mask = np.zeros(U.shape, dtype=bool)
mask[40:60, 40:60] = True
U[:20, :20] = np.nan
U = np.ma.array(U, mask=mask)

ax4 = fig.add_subplot(gs[2:, :])
ax4.streamplot(X, Y, U, V, color='r')
ax4.set_title('Streamplot with Masking')

ax4.imshow(~mask, extent=(-w, w, -w, w), alpha=0.5,
          interpolation='nearest', cmap='gray', aspect='auto')
ax4.set_aspect('equal')

plt.tight_layout()
plt.show()
```

## Barchart Demo

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple

n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

# opacity: 不透明度
opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index, means_men, bar_width,
                alpha=opacity, color='b',
                yerr=std_men, error_kw=error_config,
                label='Men')

rects2 = ax.bar(index + bar_width, means_women, bar_width,
                alpha=opacity, color='r',
                yerr=std_women, error_kw=error_config,
                label='Women')

ax.set_xlabel('Group')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('A', 'B', 'C', 'D', 'E'))
ax.legend()

fig.tight_layout()
plt.show()
```

# 3. 自分の好きな関数を３つ同じfigureでplotし、描画する

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


#fig, ax = plt.subplots(1, 3)
#ax[0].plot()t

def sin(x):
    y = np.sin(2 * np.pi * x)
    return y

def exp(x):
    y = np.exp(x)
    return y

def damped_osillation(x):
    y = np.sin(2 * np.pi * x) * np.exp(-x)
    return y

x = np.arange(-5, 5, 0.01)
y1 = sin(x)
y2 = exp(-x)
y3 = damped_osillation(x)
plt.xlim(-1, 2)
plt.ylim(-2.5, 2.5)
plt.grid(True)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='exp(-x)')
plt.plot(x, y3, label='damped_osillation(x)')
plt.legend(loc='upper right')
#plt.xticks(())
#plt.yticks(())
plt.show()
```

# 5. 自分の好きな関数を作成し,animationで動かせ
参考にしたサイト: https://sabopy.com/py/matplotlib_3d_30/

slice classについて: https://www.haya-programming.com/entry/2018/05/09/163234<br>
sliceの基本[start, stop, step]

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

def midpoints(x):
    sl = ()
    # x.ndim = 3
    # np.index_exp(ression) == np.s_
    # 配列用のインデックスタプルを作成するための良い方法
    # example:
    # >>> np.array([0, 1, 2, 3, 4])[np.s_[2::2]]
    # array([2, 4])
    #print('np.s_[:-1]: {}, np.s_[1:]:{}, np.s_[:]: {}'.format(np.s_[:-1], np.s_[1:], np.s_[:]))
    for i in range(x.ndim):
        #print('{}:sl + np.s_[:-1] {}'.format(i, sl + np.index_exp[:-1]))
        #print('{}:sl + np.s_[:-1] {}'.format(i, sl + np.index_exp[1:]))
        # 0と1の(x,y,z)、1と2(x,y,z)の点を足して２で割る（中間点の値が出る）
        # それをxとyとz軸にそれぞれずらして行っている
        x = (x[sl + np.index_exp[:-1]] + x[sl + np.index_exp[1:]]) / 2.0
        sl += np.index_exp[:]
    #print('sl[:]: {}'.format(sl))
    return x

# prepare some coordinates
# 重くなるので頂点数nを小さくした方が良い
n = 10
xr = np.linspace(-1.3,1.3,n)
yr = np.linspace(-1.3,1.3,n)
zr = np.linspace(-1.3,1.3,n)
xr, yr, zr = np.meshgrid(xr, yr, zr)
#print(xr.shape)
x = midpoints(xr)
y = midpoints(yr)
z = midpoints(zr)
#print(x.shape)
#print(x)


# 条件を満たす部分がTrue、それ以外はFalseの配列ができる
# define a heart
#heart = (x**2+(9/4)*y**2+z**2-1)**3-x**2*z**3-(9/80)*y**2*z**3 < 0
# define a square
heart = x**2+y**2+z**2 < 1

#plot
print('3D_plot')
fig = plt.figure(figsize=(5,5))
ax = fig.gca(projection='3d')
ax.voxels(xr, yr, zr, heart, facecolor="r",edgecolors='w',linewidth=0.2)
ax.set_aspect('equal')
#plt.savefig("3Dvoxel_heart.png", dpi=200,transparent = False)
plt.show()

ax.grid(False)
ax.set_axis_off()

from IPython.display import HTML
import matplotlib.animation as animation

def init():
    ax.voxels(xr, yr, zr, heart, facecolor="r",edgecolors='pink',linewidth=0.2)
    return fig,

def animate(i):
    ax.view_init(elev=30., azim=3.6*i)
    return fig,

# Animate
ani = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=70, interval=70, blit=False)    

print('Animation')
HTML(ani.to_html5_video())
```
