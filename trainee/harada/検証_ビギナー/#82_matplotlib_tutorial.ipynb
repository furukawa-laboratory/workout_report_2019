 関数gca（）は現在のAxes（matplotlib.axes.Axesインスタンス）を返し、gcf（）は現在のFigure（matplotlib.figure.Figureインスタンス）を返す

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

figure（1）はデフォルトで作成される。figure（）コマンドはオプション。
手動で軸を指定しないと、デフォルトでsubplot（111）が作成される。 
subplot（）コマンドは、numrows、numcols、plot_numberを指定する。
ここで、plot_numberは、1からnumrows * numcolsの範囲で、 numrows * numcols <10の場合、subplotコマンドのコンマはオプション。 つまりsubplot（211）はsubplot（2、1、1）に同じ。

任意の数のサブプロットと軸を作成できる。 
手動で、つまり長方形のグリッド上に軸を配置したくない場合は、axes（）コマンドを使用することで軸（[left、bottom、width、height]）として位置を指定する。 （0から1）座標 手動で軸を配置する例についてはAxes Demo、多数のサブプロットを含む例についてはBasic Subplot Demoを参照。

数字の数を増やしながらfigure（）を複数回呼び出すことで、複数のfigureを作成できる。 もちろん、それぞれの図は望むだけの数の軸とサブプロットを含むことができる。

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])
```

```python
plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default
```

```python
plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

現在のFigureはclf（）で、現在のAxesはcla（）でクリアできる。バックグラウンドで状態（特に現在の画像、図、軸）が維持されているのが厄介だと思っていても、気にしないで。 Artist tutorialを参照のこと。

たくさんのFigureを作成している場合は注意が必要で、Figureに必要なメモリは、そのFigureがclose（）で明示的に閉じられるまで完全に解放されない。 Figureへの参照をすべて削除したり、ウィンドウマネージャを使用してFigureが表示されているウィンドウを終了したりするだけでは十分ではない。pyplotはclose（）が呼び出されるまで内部参照を維持する。

text（）コマンドは任意の位置にテキストを追加するために使用でき、xlabel（）、ylabel（）およびtitle（）は指示された位置にテキストを追加するために使用される（より詳細な例はMatplotlibプロットのテキストを参照）

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density = 1, facecolor = 'g', alpha = 0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu = 100,\ \sigma = 15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

LaTeXによるテキストレンダリングを参照してください。

上記の基本的なtext（）コマンドを使用すると、Axesの任意の位置にテキストが配置される。 

テキストの一般的な用途は、プロットのいくつかの関数に注釈を付けること。

notate（）メソッドは、注釈を簡単にするためのヘルパー機能がある。 
注釈では考慮すべき点が2つあり、注釈が付けられている場所が引数xyで表されていることと、テキストxytextの場所である。 これらの引数は両方とも（x、y）タプル。

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='red', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```

この基本例では、xy（矢印の先端）とxytextの位置（テキストの位置）の両方がデータ座標にある。

## 対数および他の非線形軸

matplotlib.pyplotは、直線軸スケールだけでなく、対数スケールとロジットスケールもサポートしている。
これは、データが何桁にも及ぶ場合によく使われる。 軸のスケールを変更するのは簡単で：  
$plt.xscale('log')$

```python
from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
