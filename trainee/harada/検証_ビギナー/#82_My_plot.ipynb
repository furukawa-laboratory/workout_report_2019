```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(50)
y1 = []
y2 = []
y3 = []
for n in range(len(x)):
    yr = np.random.randn(1)
    y1.append(np.sin(yr*x[n])*2+2)
    y2.append(np.cos(yr*x[n])*2-2)
    y3.append(yr*x[n]*.1)

plt.plot(x, y1,'r:',label="y1")
plt.plot(x, y2,'bo:',label="y2")
plt.plot(x, y3,'g-',label="y3")
plt.title('three function')
plt.legend()
plt.show()
```
