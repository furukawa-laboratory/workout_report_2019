コマンドによる様々なスタイルでの描画

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dases, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

 関数gca（）は現在のAxes（matplotlib.axes.Axesインスタンス）を返し、gcf（）は現在のFigure（matplotlib.figure.Figureインスタンス）を返す

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

figure（1）はデフォルトで作成される。figure（）コマンドはオプション。
手動で軸を指定しないと、デフォルトでsubplot（111）が作成される。 
subplot（）コマンドは、numrows、numcols、plot_numberを指定する。
ここで、plot_numberは、1からnumrows * numcolsの範囲で、 numrows * numcols <10の場合、subplotコマンドのコンマはオプション。 つまりsubplot（211）はsubplot（2、1、1）に同じ。

任意の数のサブプロットと軸を作成できる。 
手動で、つまり長方形のグリッド上に軸を配置したくない場合は、axes（）コマンドを使用することで軸（[left、bottom、width、height]）として位置を指定する。 （0から1）座標 手動で軸を配置する例についてはAxes Demo、多数のサブプロットを含む例についてはBasic Subplot Demoを参照。

数字の数を増やしながらfigure（）を複数回呼び出すことで、複数のfigureを作成できる。 もちろん、それぞれの図は望むだけの数の軸とサブプロットを含むことができる。

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])
```

```python
plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default
```

```python
plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

```python
plt.figure(2)                # figure 1 current; subplot(212) still current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title

plt.figure(2)                # figure 1 current; subplot(212) still current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

現在のFigureはclf（）で、現在のAxesはcla（）でクリアできる。バックグラウンドで状態（特に現在の画像、図、軸）が維持されているのが厄介だと思っていても、気にしないで。 Artist tutorialを参照のこと。

たくさんのFigureを作成している場合は注意が必要で、Figureに必要なメモリは、そのFigureがclose（）で明示的に閉じられるまで完全に解放されない。 Figureへの参照をすべて削除したり、ウィンドウマネージャを使用してFigureが表示されているウィンドウを終了したりするだけでは十分ではない。pyplotはclose（）が呼び出されるまで内部参照を維持する。
