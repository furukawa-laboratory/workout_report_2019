```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(50)
y1 = []
y2 = []
y3 = []
for n in range(len(x)):
    yr = np.random.randn(1)
    y1.append(np.sin(yr*x[n])*2+2)
    y2.append(np.cos(yr*x[n])*2-2)
    y3.append(yr*x[n]*.1)

plt.plot(x, y1,'r:',label="y1")
plt.plot(x, y2,'bo:',label="y2")
plt.plot(x, y3,'g-',label="y3")
plt.title('three function')
plt.legend()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def func1(x, y):
    return x**3 + y**6 + 2/x + 2/y

x = np.arange(-3.0, 3.0, 0.1)
y = np.arange(-3.0, 3.0, 0.1)

X, Y = np.meshgrid(np.sin(x), np.sin(y))
Z = func1(X, Y)

fig = plt.figure(figsize=(6, 6))
ax = Axes3D(fig)
#ax.scatter3D(y1, y2, y3, c='r')
#ax.plot_wireframe(x, y1, yz)
ax.plot_wireframe(X,Y,Z,color='y')

#3D
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#ax.view_init(30, 60)
plt.show()
```
